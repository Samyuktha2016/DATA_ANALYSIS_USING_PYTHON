```python
#extracting a table from a web page
H2="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
page1=requests.get(H2)
html_code2=BeautifulSoup(page1.text,'html')
print(html_code2)
```

```python
table=html_code2.find('table')
print(table)
```

```python
th=table.find_all('th')
print(th)
```

```python
headings=[i.text for i in th]
print(headings)
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=headings)
print(mydf)
```

```python
rows=table.find_all('tr')
print(rows)
```

```python
data=table.find_all('td')
print(data)
```

```python
data1=[i.text for i in data]
print(data1)
```

```python
ind=0
start=0
stop=4
for i in data1:
  if stop<=40:
    mydf.loc[ind]=data1[start:stop]
    start+=4
    stop+=4
    ind+=1
mydf
```

```python
mydf.to_csv("Forbes.csv")
```
