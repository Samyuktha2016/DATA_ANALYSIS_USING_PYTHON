```python
#pandas
#Seies syntax--->pd.Series(col,index)
a=["Jwalitha",'ramya','durga','jahnavi','lahari','sunny','dhanush']
r=pd.Series(a, index=[67,43,44,89,34,45,23])   #pd.series   series creates as rows columns series s should be capital (col,index)
print(r)
```

```python
#seperately taking index
a=["Jwalitha",'ramya','durga','jahnavi','lahari','sunny','dhanush']
index=([67,43,44,89,34,45,23])
r=pd.Series(a,index)
print(r)
```
