```python
#Matrix multiplication by array operations by using symbol(*)
A=np.array([23,45,76,1]).reshape(2,2)
B=np.array([8,25,6,2]).reshape(2,2)
print(A)
print("___________________________")
print(B)
C=A*B   #one on one multiplication
print("___________________________")
print(C)
```

```python
#Matrix multiplication by array operations by using symbol(@)
A=np.array([23,45,76,1]).reshape(2,2)
B=np.array([8,25,6,2]).reshape(2,2)
print(A)
print("___________________________")
print(B)
C=A@B   #matrix  multiplication or matrix multiplication for multiplication @
print("___________________________")
print(C)
```

```python
#Matrix multiplication by array operations by using symbol(.-dot)
A=np.array([23,45,76,1]).reshape(2,2)
B=np.array([8,25,6,2]).reshape(2,2)
print(A)
print("___________________________")
print(B)
C=A.dot(B)   #for matrix  multiplication
print("___________________________")
print(C)
```

```python
#print the maximum and minimum of array of row and column by axis
A=np.array([34,56,78,90]).reshape(2,2)
B=np.array([7,9,3,12]).reshape(2,2)
print(A)
print(B)
print("__________________")
print(A.max(axis=0)) #axis 0---column
print("__________________")
print(B.min(axis=1)) #axis 1----row
```

```python
#Joining arrays! and printing the vertically elements in stack
A=np.array([23,45,76,1]).reshape(2,2)
B=np.array([8,25,6,2]).reshape(2,2)
print(A)
print("___________________________")
print(B)
print("After vertical stacking the array is")    #vertical stack
print(np.vstack((A,B)))
```

```python
#Joining arrays! and printing the horizantally elements in stack
A=np.array([23,45,76,1]).reshape(2,2)
B=np.array([8,25,6,2]).reshape(2,2)
print(A)
print("___________________________")
print(B)
print("After horizontal stacking the array is")   #horizontal stack
print(np.hstack((A,B)))
```

```python
# joining Arrays! and printing the vertically and horizantally elements in stack
A=np.array([34,56,78,90]).reshape(2,2)
B=np.array([7,9,3,12]).reshape(2,2)
print(A)
print("________________")
print(B)
print("after vertically stacking the arrays!")  #vertical stack
print(np.vstack((A,B)))
print("after horizontall stacking the arrays!") #horizontal stack
print(np.hstack((A,B)))
```

```python
#joining arrays stacking the elements by axis
A=np.array([23,45,76,1]).reshape(2,2)
B=np.array([8,25,6,2]).reshape(2,2)
print(A)
print("___________________________")
print(B)
print("After stacking the array is")
print(np.stack((A,B),axis=0))
```

```python
#joining arrays stacking the elements by ---> dstack
A=np.arange(30).reshape(2,3,5)   #groups,rows,columns here groups as the 2 types of 0to 29[[]]] rows as 0,5,10 and columns as 0,1,2,3,4
print(A)                         #stacks according to the dimension
print("After dstack")             #no of row == no.of groups
#no .of columns in every group == no.of rows in every group
#[0,0],[1,1],............
print(np.dstack(A))          # original   (groups,rows,columns)
                                #after dstack (columns,groups,rows)
```

SEARCHING AN ARRAY!

```python
#returns the position of the elements according to the condition  under the condition satisifies
A=np.array([12,45,2,67,23])
print(np.where(A>50)) #returns the position of the elements according to the condition
```

```python
#returns the position of the searching element in the array under the condition satisifies
A=np.array([12,45,2,67,23,45])
print(np.where(A==45))      #returns the position of the searching element
```

```python
#returns the position of the even elements in the array if the condition satisifies
A=np.array([12,45,2,67,23,45])
print(np.where(A%2==0))      #returns the position of the even elements
```

```python
#returns output the index of the seach element after sorting the list if it satisifies the condition
A=np.array([12,23,34,45,21,45])   #12 21 23 34 45 45
print(np.searchsorted(A,34))      #output the index of the seach element after sorting the list
```

# LINEAR ALGEBRA

```python
#linear algebra
A=np.array([6,34,2,67,8])
B=np.array([5,12,4,1,3])
print(np.add(A,B))                #addition of 2 arrays
print(np.subtract(A,B))           #subtraction of 2 arrays
```

```python
#quotient
print(np.divide(A,B))       #quotient
```

```python
#remainder
print(np.mod(A,B))        #remainder
```

```python
#multiplication
print(np.multiply(A,B)) #multiplication
```

```python
## both quotient and remainder(divmod)
print(np.divmod(A,B))     # both quotient and remainder
```

```python
#difference between two elements
A=np.array([6,34,2,67,8])
print(np.diff(A))      #difference between two elements
```

```python
#sort
A=np.array([6,34,2,67,8])
print(np.sort(A))
```

# SET FUNCTIONS

```python
#set functions
A=np.array([6,34,2,67,8])
B=np.array([5,12,4,1,3])
print(np.union1d(A,B))
```

```python
#Intersection
A=np.array([6,34,2,67,8])
B=np.array([5,12,4,1,2,3])
print(np.intersect1d(A,B))
```

```python
#Set difference
A=np.array([6,34,2,67,8,12])
B=np.array([5,12,4,1,3])
print(np.setdiff1d(A,B)) #it prints no common elements
```
