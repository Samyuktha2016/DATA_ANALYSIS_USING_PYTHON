```python
#Get the salary from the user if the salaryis more than or equal to 50k apply 10% tax deduction,if the salary is less than 50k apply 5% tax,if the salary is more than 70k apply 12% tax, if salary exceeds 100k print you are promoted to next level
s=int(input("enter salary"))
if s==0:
  print("enter valid salary")
elif(s<50000):
  print("Salary is: ",s-s*(0.05))
elif(s<=70000):
  print("Salary is: ",s-s*(0.1))
elif(s<=100000):
  print("Salary is: ",s-s*(0.12))
else:
  print("you are promoted")
```

# LIST COMPREHENSION

```python
#list comprehension
#newlist=[output for if] without condition
l1=[1,10,12,345,23,67,123,78,988,86]
l2=[i*i for i in l1]
print(l2)
```

```python
#list comprehension
#newlist=[output for if]
l1=[1,10,12,345,23,67,123,78,988,86]
l2=[i*i for i in l1 if i>50]
print(l2)
```

```python
#list comprehension
#newlist=[(st of if) if (cond) else (st of else) for]
l1=[1,10,12,345,23,67,123,78,988,86]
l2=[i*i if i>50 else 'x' for i in l1]
print(l2)


```

# SETS

```python
#Displaying set elements
s1={1,23,45,1,23,67,89}    #prints the unique elements in set
print(s1)
```

```python
#union
s2={1,23,1,67,89}
print(s1.union(s2))  #union--prints only the unique elements that means the repeated and non repeated elements print only once
```

```python
#Intersection
print(s2.intersection(s1))  #Intersection--prints only the intersection elements that means that only repeated elements print
```

```python
#Difference
print(s1.difference(s2))    #difference-->prints only the non repeated elements wont comethat means 45 is not repeated so it prints
```

```python
#length of a set
s={1,2,3,4,5}
print(len(s))
```

```python
#minimum and maximum of a set
print(min(s))   #minimum of a element in a set
print(max(s))   #maximum of a element in a set
```

```python
#Sum of a set
print(sum(s))  #1+2+3+4+5
```

```python
#to sort a set if a set is unsorted
s3={5,3,2,4,1}   #unordered set
print(sorted(s3))
```

```python
#updation of a set s1 into s2
s1.update(s2)
print(s1)
```

```python
#adding the element into set
s1.add(89)
print(s1)
```

```python
#removing an element in a set
s1.remove(45) #removes a particular element in a set
print(s1)
```

```python
#pop the element in a set
s1.pop()
print(s1) #removes an element in a set randomly or left or right without giving an element or index
```
